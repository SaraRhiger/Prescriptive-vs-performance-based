### Optimering naturlig ventilation så opening area holder luftskitet på under 4.5 h^-1

```python
import numpy as np
import json

from eppy.modeleditor import IDF

iddfile = "/Users/sararhiger/Desktop/V9-6-0-Energy+.idd"
IDF.setiddname(iddfile)
idfname = "/Users/sararhiger/Desktop/Optimering_Flow_NaturligVent.idf"
epwfile = "/Users/sararhiger/Desktop/Bachelor-project/DRY_2001-2010_epw_fikset.epw"

idf = IDF(idfname, epwfile)
idf1 = IDF(idfname)

#Kør simulering 
idf.run(ep_version="9-6-0")

#Importer Json fil
f=open("eplusout_hourly.json")
data = json.load(f)

alle_tidspunkter = np.array([list(x.values())[0] for x in data['Rows']]) 
airchange = (alle_tidspunkter > 4.5).any(0)
print(airchange)

while not (airchange).all():
    
    for i in range(len(airchange)):

        zone = idf1.idfobjects['ZoneVentilation:WindandStackOpenArea'][i]

        if not airchange[i]:
            zone['Opening_Area'] += 0.001

    idf1.save()
    
    idfname = "/Users/sararhiger/Desktop/Optimering_Flow_NaturligVent.idf"
    idf = IDF(idfname, epwfile)
    idf1 = IDF(idfname)
    
    # Kør simulering 
    idf.run(ep_version="9-6-0")

    # Importer Jason fil 
    f=open("eplusout_hourly.json")
    data = json.load(f)

    alle_tidspunkter = np.array([list(x.values())[0] for x in data['Rows']])
    airchange = (alle_tidspunkter > 4.5).any(0)

    print(airchange)
```
